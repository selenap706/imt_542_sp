```python
# Download necessary modules

import pandas as pd
from google.colab import files
import io
```

```python
# Access Method 1: Upload a file manually
# pros: as long as file is hosted locally, quick and easy to access
# cons: requires processing/filtering to make readable within code

# upload file manually
file_goodreads=files.upload()

# process dictionary type object into pandas dataframe
file_list = list(file_goodreads.keys())[0]
df = pd.read_csv(io.BytesIO(file_goodreads[file_list]))

# print first few rows of dataframe
print(df.head())
```

```python
# Access Method 2: API connection over HTTP for structured data to Site A
# pros: doesn't require direct interfacing or download, easy to gain access to data
# cons: doesn't use key, using key may complicate things. A more complex api would also be difficult to handle with this level of complexity for access. this is also reliant on internet connectivity

# download necessary requests library
!pip install requests
import requests

# define chosen API
url = "https://catfact.ninja/fact"

# request data
response = requests.get(url)

# check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse the response as JSON
    print("Random Cat Fact:", data['fact'])  # Print the fact from the response
else:
    print("Failed to retrieve data. Status code:", response.status_code)
```

```python
# Access Method 3: HTTP to download webpage
# pros: file is in html format, which is quickly readable and transportable to other formats in text
# cons: can not access images/audio/video website may have, is difficult to manage with dynamic content

# define URL for Vancouver, WA library
url = "https://www.fvrl.org/loc/vancouver"

# send an HTTP GET request to the server
response = requests.get(url)

# check if the request was successful
if response.status_code == 200:
    # save the HTML content of the webpage to a file
    with open("downloaded_webpage.html", "w", encoding="utf-8") as file:
        file.write(response.text)

    print("Webpage downloaded successfully!")

    # open the file in read mode to display its content
    with open("downloaded_webpage.html", "r", encoding="utf-8") as file:
        content = file.read()
        print(content)
else:
    print("Failed to download webpage. Status code:", response.status_code)
```
